```python
glasclassifierGNB = GaussianNB()
glasclassifierGNB.fit(glasX_train, glasY_train)

glasY_predGNB = glasclassifierGNB.predict(glasX_test)

print(classification_report(glasY_test, glasY_predGNB))
print(confusion_matrix(glasY_test, glasY_predGNB))
print('accuracy is',accuracy_score(glasY_predGNB,glasY_test))
```
